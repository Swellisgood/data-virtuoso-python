```python
# Read in the CSV
import pandas as pd
office_episodes = pd.read_csv("https://raw.githubusercontent.com/Swellisgood/pythonista-data-science-ninja/main/Netflix-investigations/netflix-unguided-project/datasets/office_episodes.csv")

# Print the first ten rows of the DataFrame
office_episodes[:10]
```

```python
# Create a color_scheme list
colors = []

# Iterate over rows of office_episodes to input color name to the colors list
for lab, row in office_episodes.iterrows():
    if row['scaled_ratings'] < 0.25:
        colors.append("red")
    elif 0.25 <= row['scaled_ratings'] < 0.50:
        colors.append("orange")
    elif 0.50 <= row['scaled_ratings'] < 0.75:
        colors.append("lightgreen")
    else:
        colors.append("darkgreen")

# Inspect the first 10 values in the list      
colors[:10]
```

```python
# Sizing system:
# episodes with guest appearances have a marker size of 250
# episodes without are sized 25

sizes = []

for lab, row in office_episodes.iterrows():
    if row['has_guests'] == True:
        sizes.append(250)
    else:
        sizes.append(25)

# Inspect the first 10 values in the list      
sizes[:10]
```

```python
# add colors and sizes columns to the DataFrame
office_episodes['colors'] = colors
office_episodes['sizes'] = sizes
```

```python
# splitting DataFrame
guests_df = office_episodes[office_episodes['has_guests'] == True]
no_guests_df = office_episodes[office_episodes['has_guests'] == False]
```

```python
# Import matplotlib.pyplot under its usual alias and create a figure
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(11,7))

# scatter plot of episode_number and viewership_mil in no_guests_df
plt.scatter(no_guests_df['episode_number'], 
            no_guests_df['viewership_mil'], 
            c=no_guests_df['colors'], 
            s=no_guests_df['sizes'], 
            alpha=0.8
           )

# scatter plot of episode_number and viewership_mil in guests_df
plt.scatter(guests_df['episode_number'], 
            guests_df['viewership_mil'], 
            c=guests_df['colors'], 
            s=guests_df['sizes'], 
            alpha=0.8, 
            marker='*'
           )

# Create a title
plt.title('Popularity, Quality, and Guest Appearances on the Office', size = 16, fontweight='bold')

# Create an x-axis and an y-axis
plt.xlabel('Episode Number', size = 15)
plt.ylabel('Viewership (Millions)', size = 15)

# Show the plot
plt.show()
```

```python
#highest number of views
highest_view = max(office_episodes["viewership_mil"])

# Filter Dataframe row of the most watched episode
most_watched_dataframe = office_episodes.loc[office_episodes["viewership_mil"] == highest_view]

# Top guest stars in that episode
top_stars = most_watched_dataframe[["guest_stars"]]
top_stars 
```
